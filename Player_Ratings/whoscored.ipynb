```python
from selenium import webdriver
import pandas as pd
import numpy as np
```

```python
driver_path = "/Users/mmac/Desktop/PythonWorkspace2/chromedriver"
driver = webdriver.Chrome(executable_path=driver_path)
```

```python
url = 'https://www.whoscored.com/Statistics'
#url = "http://www.google.com/"
driver.get(url)
```

```python
def iteration_scrapping():
    """
    Xpath is very sensitive to current status of a website.
    Before running the code, each Xpath needs to be recalibrated
    
    """
    for i in range(1,11):
        rate_element = driver.find_elements_by_xpath('//*[@id="player-table-statistics-body"]/tr['+str(i)+']/td[13]')
        rate = [x.text for x in rate_element]
        rate_list.append(rate[0])
        player_element = driver.find_elements_by_xpath('/html/body/div[5]/div[3]/div[5]/div[1]/div[3]/div/table/tbody/tr['+str(i)+']/td[1]/a[1]')
        player = [x.text for x in player_element]
        player_list.append(player[0])

#Create 2 lists for ratings and the players        
rate_list = []
player_list = []

#Click on all players tab. This part finds and clicks on a specific div inside the page
all_players=driver.find_element_by_xpath('//*[@id="apps"]/dd[2]/a')
all_players.click()

#Scrape first page. This is a task specific code, it may not be necessary in different projects.
iteration_scrapping()

next_page=driver.find_element_by_xpath('/html/body/div[5]/div[3]/div[5]/div[1]/div[4]/div/dl[2]/dd[3]/a')
next_page.click() 

iteration_scrapping()

#Scrape the rest of the pages (233). Each page has 10 players.
#Sleep s necessary to wait until the page full loaded, otherwise an error will occur
import time

for i in range(1,234): 
    print(i)
    time.sleep(5)
    #click on next page option
    next_page=driver.find_element_by_xpath('/html/body/div[5]/div[3]/div[5]/div[1]/div[4]/div/dl[2]/dd[3]/a')
    next_page.click() 
    iteration_scrapping()
```

```python
#A couple of simple data wrangling 
df = pd.DataFrame([player_list,rate_list]).T

df.drop_duplicates(subset=None, keep='first', inplace=True)

df.iloc[1944:,:][0] = df.iloc[1944:,:][0].apply(lambda row : row[5:]) 

df = df.reset_index().drop('index', axis = 1)\
                .rename(columns = {0:'Player', 1:'Rating'})
```
